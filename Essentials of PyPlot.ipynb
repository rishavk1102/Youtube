```python
def graph(t):
    return np.cos(2 * np.pi * t) * np.exp(-t)

t1 = np.arange(1, 10, 0.1)
t2 = np.arange(1, 5, 0.2)

plt.figure(1) # Creates a figure

plt.subplot(2,1,1) # 2 rows 1 column and currently talking about 1st sub-plot

plt.grid()
plt.plot(t1, graph(t1), 'r--')

plt.subplot(2,1,2)

plt.grid()
plt.plot(t2, graph(t2), 'b--')

plt.show()
```
